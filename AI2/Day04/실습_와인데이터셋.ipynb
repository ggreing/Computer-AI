```python
#와인데이터셋 데이터 분석, 전처리 과정-시각화, 모델 생성/학습/예측, 정확도확인
#교차검증(3가지 중 2가지) 정확도 확인

from sklearn.datasets import load_wine

wine = load_wine()
wine
wine.keys()
wine.feature_names
features = wine.data
label = wine.target
# 알콜
# 말산
# 회분
# 회분의 알칼리도
# 마그네슘
# 총 폴리페놀
# 플라보노이드 폴리페놀
# 비 플라보노이드 폴리페놀
# 프로안토시아닌
# 색상의 강도
# 색상
# 희석 와인 od280/od315 비율
# 프롤린
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsClassifier

dt = DecisionTreeClassifier()
dt
knn = KNeighborsClassifier(n_neighbors = 5)
knn
```

```python
kfold = KFold(n_splits = 45)

cv = []
print(features.shape)


for train_index, test_index in kfold.split(features):
    print(f"train index = {train_index}\ntest index = {test_index}")
```

```python
n_iter = 0 #반복횟수를 나타내는 변수

for train_index, test_index in kfold.split(features):
    #kfold.split(features) : 학습 데이터와 검증 데이터의 인덱스
    #train_index : 학습 데이터의 인덱스 배열
    #test_index : 검증 데이터의 인덱스 배열
    
    #데이터 분할
    X_train, X_test = features[train_index], features[test_index]
    y_train, y_test = label[train_index], label[test_index]
    
    #학습
    dt.fit(X_train, y_train)
    
    #예측
    y_pred = dt.predict(X_test)
    n_iter += 1
    
    #반복시마다 정확도 측정
    acc = np.round(accuracy_score(y_test, y_pred), 5)
    train_size = X_train.shape[0]
    test_size = X_test.shape[0]
    print(f"{n_iter} 교차검증 정확도 : {acc}, 학습데이터 크기 : {train_size}, 검증데이터 크기 : {test_size}")
    print(f"{n_iter} 검증 세트 인덱스 : {test_index}")
    cv.append(acc)
    
print(cv)

# 개별 정확도를 합하여 평균 정확도 계산
print(f"평균 검증 정확도 : {np.mean(cv)}")
```
