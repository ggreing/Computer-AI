# 성별 컬럼 레이블 인코딩

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# 레이블 인코더 생성
le = LabelEncoder()
le
```

```python
# 성별 열을 레이블 인코딩
df['Sex_LE'] = le.fit_transform(df['Sex'])

# 인코딩 클래스 조회
print(le.classes_)
```

```python
df.head()
```

```python
def encode_features(df):
    features = ['Sex', 'Cabin', 'Embarked']
    for i in features:
        le = LabelEncoder()
        le = le.fit(df[i])
        df[i] = le.transform(df[i])
        
    return df

df1 = encode_features(df)
df1

df1.info()
```
