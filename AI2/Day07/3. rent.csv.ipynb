```python
# 상관관계의 값은 -1 ~ 1까지, 0에 가까울수록 상관관계 약하거나 없음
#1. 임대료와 BHK 상관관계
print(f"BHK와 임대료 상관관계 : {df['BHK'].corr(df['Rent']):.2f}")

# 2, 임대료와 size 상관관계
print(f"BHK와 임대료 상관관계 : {df['Size'].corr(df['Rent']):.2f}")

# 3. 임대료와 도시별 상관관계 => 도시 이름으로 되어있으므로 수치형으로 변경해야함
# 각 도시의 평균 임대료 구하고 평균 임대료간의 상관관계 확인
cities = df['City'].unique()
print(cities)

city_mean = df.groupby('City')['Rent'].mean()
city_mean

print(f"임대료와 도시별 평균 임대료의 상관관계 : {df['Rent'].corr(city_mean)}")
# .corr() 각 열 간의 상관계수를 반환하는 메소드

df['City_Mean'] = df.groupby('City')['Rent'].transform('mean')
# groupby('City') : 도시별 그룹화
# groupby('City')['Rent'] : 그룹화된 각 도시 데이터에 대해 Rent 열 선택
# .transform('mean') : 선택된 임대료 데이터에 대해 평균을 계산,
# 원래 데이터프레임의 인덱스 순서를 유지한 채로 평균값을 반환
df
print(f"임대료와 도시별 평균 임대료의 상관관계 : {df['Rent'].corr(df['City_Mean']):.2f}")
```

```python
rent_city = df.groupby('City')['Rent'].corr(df['Rent'])
print(rent_city)
# 도시별 Rent 열과 Rent 열 간의 상관관계가 1.0
# 모든 도시에서 Rent열과 Rent 사이에 완벽한 선형 상관관계가 있다는 것을 의미
```

```python
# 도시 목록 가져오기
cities = df['City'].unique()
print(cities)

# 도시별 임대료 상관관계 계산
city_rent = {}
for i in cities:
    city_data = df[df['City'] == i]
    correlation = city_data['Rent'].corr(city_data['Rent'])
    city_rent[i] = correlation
    
print(city_rent)
    
```

```python
# Rent/Size 평당가 도시와의 상관관계
# 평당 가격 계산
df['Price_Per'] = df['Rent'] / df['Size']
df

# 도시목록 가져오기
print(cities)

# 도시별 평당가격과 임대료의 상관관계
city_price = {}

for i in cities:
#     print(i)
    city_data = df[df['City'] == i]
    correlation = city_data['Price_Per'].corr(city_data['Rent'])
    city_price[i] = correlation
print(city_price)
#-1~1 사이의 범위를 갖는다
#양의 값은 양의 선형 관계
#음의 값은 음의 선형 관계
#0에 가까운 값은 선형관계가 거의 없음

# 'Kolkata': 0.4300772768773486, -> 약한 양의 상관관계있음
# 'Mumbai': 0.7665554308801846, -> 강한 양의 상관관계있음
# 'Bangalore': 0.9747580556685503,  -> 매우 강한 상관관계있음
# 'Delhi': 0.21085116743392723, -> 매우 약한 양의 상관관계있음
# 'Chennai': 0.754641764525127,  -> 강한 양의 상관관계있음
# 'Hyderabad': 0.1133549904153024 -> 매우 약한 양의 상관관계있음

#임대료가 증가할 때 해당 도시의 평당가도 증가한다는 것을 시사하고 있다
#상관관계는 인과관계를 나타내지 않는다. 상관관계가 높다고 해서 두 변수 사이에
#인과관계가 있는 것은 아니다
```
